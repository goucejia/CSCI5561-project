```python
# name:keypoints_y:keypoints_x

# name = '010_360_001'
def process_single_file(name, text_path):
  # txt_path = name + ".txt"
  with open(text_path) as f:
    lines = f.readlines()
    x_arr = []
    y_arr = []
    for line in lines:
      sp = line.split(",")
      # print(sp)
      x_y = sp[0].split(" ")
      # print(x_y)
      x = x_y[0]
      y = x_y[1].replace("\n", "")

      if x == "0.000000":
        x = "-1"
      if y == "0.000000":
        y = "-1"
      # print(x)
      # print(y)
      y_arr.append(float(y))
      x_arr.append(float(x))

    # update_annotate = name+".jpg"+":["+(", ".join(y_arr))+"]:["+(", ".join(x_arr))+"]"
    update_annotate = [name+".jpg", y_arr, x_arr]
    # print(update_annotate)
    return update_annotate
  
```

```python
import os
import csv

def gen_img_pairs(in_path, out_csv_path):
  file_names = []
  for root, dirs, files in os.walk(in_path):
    for name in files:
      file_names.append(name)
  print(file_names)

  with open(out_csv_path, mode='w') as target_file_csv:
    csvwriter = csv.writer(target_file_csv, delimiter=',')
    csvwriter.writerow(["", "from", "to"])
    iter = (len(file_names) // 20)
    for i in range(iter):
      # cur_str = str(i)+","+str(file_names[i])+","+str(file_names[i+1])+",,,,,"
      for j in range(19):
        cur_index = i*20+j
        cur_str = [cur_index, file_names[cur_index], file_names[cur_index+1]]
        # print(cur_str)
        csvwriter.writerow(cur_str)
```

```python
in_subject_path = "/content/gdrive/MyDrive/5561_project/training_data/Training_set/subject_120"
out_subject_path = "/content/gdrive/MyDrive/5561_project/subject_converted"
files = os.listdir(in_subject_path)
i = 0
for f in files:
  # print(f)
  in_kp_path = in_subject_path+"/"+f+"/image"
  print(in_kp_path)
 
  out_kp_path = out_subject_path+"/subject_120/"+f+"-pairs.csv"
  print(out_kp_path)

  gen_img_pairs(in_path=in_kp_path, out_csv_path=out_kp_path)
  

  print()
  i += 1

print("processed: {}".format(i))
```
